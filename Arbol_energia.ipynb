```python
import pandas as pd 
from sklearn.model_selection import train_test_split # metodo train_test_split  que ayuda a generar subconjuntos de entrem,aiento y prueba para el arbol


data= pd.read_csv("iris.csv")
X=data.drop ("variety", axis=1) #la x siempre es mayuscula y con esta liena se elimina especies
y= data["variety"] # clase


```

```python
#generar conjunto de entremanieto y conjunto de prueba (obtiene la presione  y etc )

X_train , X_test, y_train, y_test = train_test_split(
    X,y,
    test_size= 0.3,
    random_state= 42, #parametro que ayuda a que todas las ejecuciones sean iguales se define algo llamado semilla que ayuda a obtener el mismo resultado
    stratify= y # es igual a y , quiere decir que quiere tener una variable para que conserve la proporci√≥n separa el conjunto de entremiento y prueba 
)
```

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(
    criterion="entropy",
    random_state=42
)
```

```python
model.fit(X_train, y_train)
# Visualizaci√≥n
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=X.columns,
class_names=y.unique(), filled=True)
plt.show()
```

```python
#Despues del arbol queremos saber que tan preciso es el modelo 
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = model.predict(X_test) # el modelo hace la predciion con los atributos que estan en la prueba 
print("Matrizde confusi√≥n:\n", confusion_matrix(y_test, y_pred)) # La matriz de confusion se crea con esta linea 
print("Accuracy:", accuracy_score(y_test, y_pred)) 

#Accuacy es la predicci√≥n o exactitud buena precisi√≥n 70%
```

```python
#la validaci√≥n cruzada es un atecnica que permite generar modelos y evaluarlos con pocos datos  se tiene que tener por lo mennos 30 ejemplos 
#valizaci√≥n de 3 pliegues por que se va dividir en 3 partes  cada pliegeue va generar 3 modelos (modelo 1 ocupa entremiento 2, 3 prueba 1 ), (modelo 2 ocupa entremiento 1, 3 prueba 2 ) y (modelo 3 ocupa entremiento 1,2 prueba 3 )

from sklearn.model_selection import cross_val_score 
scores = cross_val_score(model, X, y, cv=10, scoring="accuracy") # cv=10 es la cantidad de pliegues 
print("Accuracy promedio:", scores.mean()) # con este se obtiene el desempe√±o promedio 
#decargar conjunto de open ml
#estudiar de la pagina 63 
```
