```python
import pandas as pd
df = pd.read_csv("Occupancy_Estimation_continuo.csv")
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Room_Occupancy_Count", "Date","Time"], axis=1) #axis 1 refiere a eliminar columnas, se eliminara la de Room_Occupancy_Count
y = df["Room_Occupancy_Count"]

```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
test_size=0.2, #Pocos datos .20 esta bien, muchos datos .50 mejor, 30 prueba 70 entrenamiento
random_state = 42,
stratify = y) #si no dejamos el 42 cuando ejecutemos todos los resultados seran distintos, omitir o documentar, guardas la semilla de los resultados que obtuviste en ese 42 para poder duplicar 
# o volver a obtener este resultado
```

```python

X_test #cross validation
#.shape da las dimensiones , cuando hagamos pruebas asegurarse de tener por lo menos 30 ejemplos en el conjunto de prueba. (Solo en el test)
```

```python
#Hacer el modelo, no hay que hacer getdumies porque los datos son numericos 
from sklearn.neighbors import KNeighborsClassifier #Vamos a clasificar primero
knn_clf = KNeighborsClassifier(
    n_neighbors =10,
    weights= "uniform" #que tanta importancia le vamos a dar a los puntos en nuestro conjunto de datos, con uniform nos referimos a que todo
    #vale lo mismo
)
```

```python
knn_clf.fit(X = X_train, y = y_train) #usamos los datos de entrenamiento
```

```python
knn_clf.predict(X_test) #Predecir / clasificar
```

```python
y_pred = knn_clf.predict(X_test) 
```

```python
#Ahora evaluaremos 
from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_true=y_test, y_pred=y_pred))
# se equivoco en 2
```

```python
print(accuracy_score(y_true=y_test, y_pred=y_pred)) # y_true, y_test, el primer nombre es el nombre de ese atributo y el 2 que metemos osea 
#y_test y y_pred, ya que son los segundos son los que metemos.  
```
